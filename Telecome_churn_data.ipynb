# scaling

```python
#It should be performed on numerical data

from sklearn.preprocessing import StandardScaler

```

```python
tele_data
```

```python
excluded_cols = ["gender","mail_subscribed","multi_screen","churn"]
scale_col = [col for col in tele_data.columns if col not in excluded_cols]


scaler = StandardScaler()
scaled_array = scaler.fit_transform(tele_data[scale_col])
scaled_df = pd.DataFrame(scaled_array, columns=scale_col, index=tele_data.index)

encoded_df = tele_data[excluded_cols].reset_index(drop=True)
final_tele_data= pd.concat([scaled_df, encoded_df], axis=1)
```

# confusion matrix

```python
import numpy as np

tp = np.sum((y_test == 1) & (y_pred == 1)) #tn
tn = np.sum((y_test == 0) & (y_pred == 0))  # tp
fp = np.sum((y_test == 0) & (y_pred == 1))  # fp
fn = np.sum((y_test == 1) & (y_pred == 0)) #fn

print(f"tp:{tp}, tn {tn}, fp: {fp}, rn: {fn}")
```

```python
accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp) 
recall = tp / (tp + fn) 
f1_score = 2 * precision * recall / (precision + recall) 
accuracy,precision,recall,f1_score
```
